Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # Also for plotting
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing and creating an instance of the Califorrnia House Price Dataset from sklearn.datasets

```python
house_price_dataset = sklearn.datasets.fetch_california_housing()
```

```python
print(house_price_dataset)
```

```python
# Loading the dataset to a Pandas DataFrame
house_price_dataframe = pd.DataFrame(house_price_dataset.data, columns = house_price_dataset.feature_names)
```

```python
# Print First 5 rows of our DataFrame
house_price_dataframe.head()
```

```python
# add the target (price) column to the DataFrame
house_price_dataframe['price'] = house_price_dataset.target
```

```python
house_price_dataframe.head()
```

```python
# checking the number of rows and Columns in the data frame
house_price_dataframe.shape
```

```python
# check for missing values
house_price_dataframe.isnull().sum()
```

```python
# statistical measures of the dataset
house_price_dataframe.describe()
```

Understanding the correlation between various features in the dataset

There are **2 types of correlation**.

1. ***Positive Correlation-*** If value of one variable increases, then the value of the second variable also increases.

MAX VALUE FOR CORRELATION = 1 means both the values are equal so if one increases then second increases too. The diagonal of a heat map is 1 and darkest in the colour representing positive correlation.

Lighter colours shows negative correlation. E.g., in the map below the white square show as latitude increases then longitude decreases.

2. ***Negative Correlation-*** If value of one variable increases, then the value of the second variable decreases.

What a heat map does is essentially takes all the features on the X and Y Axis and checks how they are correlated with one another. The Diagonal has features checking themselves against each other so is of least significance to us.

```python
correlation = house_price_dataframe.corr()
```

```python
# constructing a heatmap to understand the correlation
# So essentially we plot the heat map for the values in correlation.
# cbar=True to show the colour bar on the side of the graph with positive and negative values showing positive and negative correlation.
# square=True to show all the values in a square
# fmt='.1f' to have only 1 value after the decimal point. It asks how many float values do u want after the decimal?
# annot=True to show annotions of the features in the dataframe in the graph
# annot_kws={'size':10} to set the size of the numbers in the graph
# cmap='Blues' to decide the colour of the heat map

plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':10}, cmap='Blues')
```

Splitting the data and Target (Price) to feed our ML algorithm in order for it to find the pattern between the price and all the other features and eventually learn from the data.

```python
# X has all the feature labels except price
# Y has all the target (price) labels
# axis=1 for dropping a column and axis=0 for dropping a row
X = house_price_dataframe.drop(['price'], axis=1)
Y = house_price_dataframe['price']
```

```python
print(X)
print(Y)
```

Splitting the data into Training data and Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

***Training XGBoost Regressor Model-***
Its a decision tree based ensemble model. Ensemble meaning more than 1 model is used together. Used this model since the dataset isn't too big

```python
# creating  the model instance
model = XGBRegressor()
```

```python
# We fit the model with X_train data which is essentially training the model
model.fit(X_train, Y_train)
```

***Evaluating the model-***
We use accuracy score as the evalutaion metric for classification problems but since this is a regression problem and all these are numerical values we use r^2 error and mean absolute error

Prediction on training data

Since the model is trained on X_train data it should be able to predict the Y_train values.

Once predicted we check the error by comparing the prediction of the model with original values.

```python
# accuracy for prediction on training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_train, training_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

# More data and other models can reduce the error
print("R squared error : ", score_1) # Should be close to 0. If its 10+ then model isn't performing well
print('Mean Absolute Error : ', score_2) # Should be close to 0. If its 5+ then model isn't performing well
```

Visualizing the actual Prices and predicted prices


Prediction on Training Data

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Predicted Price")
plt.show()
```

Prediction on Test Data

```python
# accuracy for prediction on test data
test_data_prediction = model.predict(X_test)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_test, test_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print("R squared error : ", score_1)
print('Mean Absolute Error : ', score_2)
```

```python
plt.scatter(Y_test, test_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Predicted Price")
plt.show()
```
